```python
# Question 1. What is hypothesis testing in statistics ? 
''' Ans=> Hypothesis testing is a statistical procedure used to test assumptions or hypotheses about a population parameter. It involves formulating a 
        null hypothesis (H0) and an alternative hypothesis (Ha), collecting data, and determining whether the null hypothesis can be rejected.
       Hypothesis testing is a form of statistical inference that uses data from a sample to draw conclusions about a population parameter or a population probability distribution.
       This process helps researchers verify the validity of their findings and make informed decisions based on the data collected.'''
```

```python
# Question 2. What is the null hypothesis, and how does it differ from the alternative hypothesis ? 
''' Ans=> The null hypothesis (H0) is a statement that suggests there is no relationship between two variables or no effect in the population.
 It is typically what researchers aim to disprove or reject through their experiments.
 On the other hand, the alternative hypothesis (Ha) is a statement that claims there is a statistical relationship between the two variables or an 
    effect in the population.
 Researchers generally aim to accept or prove the alternative hypothesis.
 The null hypothesis is denoted by H0 and uses an equality symbol ( =, >=, <=), while the alternative hypothesis is denoted by H1 or Ha and uses an
    inequality symbol ( !=, <, >).'''
```

```python
# Question 3. What is the significance level in hypothesis testing, and why is it important ? 
''' Ans=> The significance level in hypothesis testing is the probability set by the researcher to determine the threshold at which the null hypothesis 
           is either accepted or rejected.
         It represents the likelihood of making a type I error and is typically set at 1%, 5%, or 10% based on various factors such as research type, 
           sample size, and associated costs.
         The significance level is denoted by the Greek letter alpha (α) and is the fixed probability of wrong elimination of the null hypothesis when 
           it is true.
        
           It is important because it helps researchers decide whether the results of their study are statistically significant.
                 
```

```python
# Question 4. What does a P-value represent in hypothesis testing ? 
''' Ans=> A P-value in hypothesis testing represents the probability of obtaining test results at least as extreme as the result actually observed, under
            the assumption that the null hypothesis is correct.
         It quantifies the evidence against the null hypothesis, with a low P-value suggesting that the data is inconsistent with the null hypothesis, 
            potentially favoring an alternative hypothesis.
         The P-value does not measure the probability that the studied hypothesis is true or the probability that the data were produced by random 
            chance alone.
```

```python
# Questiom 5.  How do you interpret the P-value in hypothesis testing ?
''' Ans=> The P-value in hypothesis testing is the probability of obtaining test results at least as extreme as the result actually observed, under the
             assumption that the null hypothesis is correct.
             It quantifies the evidence against the null hypothesis. A low p-value suggests data is inconsistent with the null, potentially favoring an
               alternative hypothesis.
            
            To interpret the p-value, you need to compare it to a chosen significance level (α), generally 5% (α = 0.05). If the p-value is less than or
                equal to α, you reject the null hypothesis, indicating that the observed effect or relationship is statistically significant.
             Conversely, if the p-value is greater than α, you fail to reject the null hypothesis, meaning the sample data does not provide strong
                 enough evidence to rule out the possibility that the effect is due to chance.
            
            It is important to note that a p-value does not measure the probability that the studied hypothesis is true, or the probability that the data
                were produced by random chance alone.
             Additionally, rejecting the null hypothesis does not prove the alternative hypothesis; it just suggests that the alternative hypothesis may
               be plausible given the observed data.
```

```python
# Question 6. What are Type 1 and Type 2 errors in hypothesis testing ? 
''' Ans=> Type 1 and Type 2 errors are incorrect conclusions that can occur in hypothesis testing. 
        1. Type 1 error
          A Type 1 error, also known as a false positive, occurs when a true null hypothesis is incorrectly rejected.
         This means that you conclude there is a significant effect or difference when there isn't one.
         For example, in a courtroom scenario, a Type 1 error would be convicting an innocent defendant.
        
        2. Typr 2 error
         Type 2 error, also known as a false negative, occurs when a false null hypothesis is not rejected.
         This means that you fail to detect a significant effect or difference when one actually exists.
         In the courtroom example, a Type 2 error would be acquitting a criminal.
        
        Both types of errors are important to consider in statistical analysis, as they can lead to incorrect decisions and interpretations of data.
```

```python
# Question 8 . What is the Z-test, and when is it used in hypothesis testing. 
''' Ans=> The Z-test is a statistical test used to determine if there is a significant difference between the means of two populations. It is based on 
             the assumption that the data follows a normal distribution and is particularly useful when the population variance is known or when the 
               sample size is large (typically 30 or more data points).
         The Z-test compares the sample mean to the population mean and calculates a Z-score, which indicates how many standard deviations the sample
             mean is from the population mean.
        
        The Z-test is used in hypothesis testing when the population standard deviation is known and the sample size is sufficiently large, allowing the
               use of the normal distribution to approximate the sampling distribution of the test statistic.
         It is also used when comparing proportions between two independent groups, provided the sample sizes are large enough for the Central Limit Theorem to apply.
```

```python
# Question 9. How do you calculate the Z-score, and what does it represent in hypothesis testing ? 

''' Ans=> The Z-score is calculated by subtracting the population mean from the observed value and dividing the result by the population standard deviation.
         This can be represented by the formula: Z= X−μ / σ
         , where X is the observed value, μ is the population mean, and σ is the population standard deviation.
        
        In hypothesis testing, the Z-score represents how many standard deviations an observed value is from the mean of the population. It is used to 
           determine the probability of observing a particular value under the assumption that the null hypothesis is true. By comparing the Z-score to 
             critical values from the standard normal distribution, researchers can decide whether to reject or fail to reject the null hypothesis
```

```python
# Question 10. What is the T-distribution, and when should it be used instead of the normal distribution ? 
''' Ans=> The t-distribution, also known as Student's t-distribution, is a probability distribution that is similar to the normal distribution but has 
             heavier tails, which means it gives more weight to extreme outcomes. This makes it more suitable for situations where there is greater 
               uncertainty, such as when dealing with small sample sizes or when the population standard deviation is unknown.
         The t-distribution is defined by the degrees of freedom, which are related to the sample size.
        
        The t-distribution should be used instead of the normal distribution when estimating the mean of a normally distributed population with small 
          sample sizes and an unknown population standard deviation.
         It is particularly useful in scenarios where the sample size is small, and the population variance is not known, as it provides a more accurate
           estimate of the population parameters.'''
```

```python
# Question 11. What is the difference between a Z-test and a T-test ? 
''' Ans=>
Z Test and T- test
A Z-test and a T-test are statistical hypothesis tests used to compare means, but they differ in several key aspects. A Z-test is typically used when
  the sample size is large (n ≥ 30) and the population standard deviation is known.
 In contrast, a T-test is used when the sample size is small (n < 30) and the population standard deviation is unknown.

The Z-test assumes that the data follows a normal distribution and uses the standard normal distribution (Z-distribution) to calculate the test statistic.
 
 On the other hand, the T-test uses the T-distribution, which accounts for the additional uncertainty due to estimating the population standard deviation from the sample data.

the T-test requires the calculation of degrees of freedom, which is not applicable in the Z-test.
 The Z-test is generally simpler to compute, while the T-test is more complex due to the need to estimate the standard deviation.

Both tests can be used for one-sample or two-sample comparisons, but the Z-test is more commonly used for two-sample comparisons.
 The choice between a Z-test and a T-test depends on the sample size and whether the population standard deviation is known. '''
```

```python
# Question 12. What is the T-test, and how is it used in hypothesis testing ? 
''' Ans=> The T-test is a statistical hypothesis test that assesses sample means to draw conclusions about population means.
         It is commonly used to determine if there is a significant difference between the means of two groups. The T-test is particularly useful when the 
           population variance is unknown and the sample size is small.
        
        There are different types of T-tests, including the one-sample T-test, which evaluates whether the mean of a single group is significantly different from 
           a specified value, and the two-sample T-test, which compares the means of two independent groups.
         Another type is the paired T-test, which is used when the same subjects are measured twice, such as before and after an intervention.
        
        In hypothesis testing, the T-test calculates a T-statistic, which is then compared to a critical value from the T-distribution to determine if the null
          hypothesis can be rejected. '''
```

```python
# Question 13. What is the relationship between Z-test and T-test in hypothesis testing ? 
''' Ans=> The Z-test and T-test are both statistical hypothesis tests used to compare means, but they differ in their assumptions and applications. Both
           tests are parametric and require raw data, calculate test statistics (Z-statistic for Z-test and T-statistic for T-test), and determine p-values
            to assess the significance of observed differences.
         However, the Z-test is typically used when the sample size is large (n ≥ 30) and the population variance is known, while the T-test is used for 
           small samples (n < 30) or when the population variance is unknown.
         The T-test is more flexible and reliable in dynamic environments, such as online experiments, where user behavior can be unpredictable.
         Additionally, the T-test uses the Student’s t-distribution, which has thicker tails than the standard normal distribution used by the Z-test,
           making it more suitable for smaller sample sizes.
         When the sample size is large, the t-distribution converges to the standard normal distribution, making the Z-test and T-test results similar.
         However, in practice, the T-test is often preferred when the population variance is unknown, regardless of sample size, due to its robustness.'''
```

```python
# Question 14. What is a confidence interval, and how is it used to interpret statistical results ? 
''' Ans=> A confidence interval (CI) is a range of values used to estimate an unknown statistical parameter, such as a population mean. It provides a
           range, such as 2 to 4 hours, along with a specified confidence level, typically 95%. This indicates that if the same sampling procedure were
             repeated 100 times, approximately 95 of the resulting intervals would be expected to contain the true population mean.
         Confidence intervals describe the variation around a statistical estimate and predict what the value of your estimate is likely to be.
         When you make an estimate in statistics, there is always uncertainty around that estimate because the number is based on a sample of the 
           population you are studying. The confidence interval is the range of values that you expect your estimate to fall between a certain percentage
            of the time if you run your experiment again or re-sample the population in the same way.'''
```

```python
# Question 15. What is the margin of error, and how does it affect the confidence interval ? 
''' Ans=> The margin of error is a statistic that expresses the amount of random sampling error in the results of a survey or study. It indicates how many
            percentage points the results will differ from the true population value.
         The margin of error is calculated by multiplying a critical value (based on the desired confidence level) by the standard error of the estimate.
         For example, if a survey reports a 55% approval rating with a 5% margin of error, the true approval rating is expected to be between 50% and 60%.
        
        The margin of error directly affects the confidence interval. The confidence interval is a range of values that is likely to contain the true 
          population parameter, and the margin of error represents half the width of this interval.
         A larger margin of error results in a wider confidence interval, indicating less precision in the estimate. Conversely, a smaller margin of 
           error leads to a narrower confidence interval, suggesting greater precision.'''
```

```python
# Question 16. How is Bayes' Theorem used in statistics, and what is its significance ?
''' Ans=> Bayes' Theorem is a fundamental concept in statistics that allows for the calculation of conditional probabilities, enabling the updating of 
            probabilities based on new evidence or information.
          It provides a mathematical framework to relate the probability of an event given some prior knowledge to the probability of the prior knowledge
             given the event.
         This theorem is particularly useful in situations where we need to revise our beliefs or predictions based on additional data, making it a 
            cornerstone of Bayesian statistics.
            
          In statistical inference, Bayes' Theorem is used to update the probability of a hypothesis as more evidence or information becomes available. 
             This is done by combining the prior probability of the hypothesis with the likelihood of the observed data under the hypothesis, resulting 
                in a posterior probability that reflects the updated belief about the hypothesis.
          The theorem is also applied in various fields such as medicine, finance, and machine learning, where it helps in making more accurate predictions 
             and decisions by incorporating prior knowledge and new data.
            
            The significance of Bayes' Theorem lies in its ability to handle uncertainty and provide a systematic way to incorporate new evidence into 
               existing knowledge. It is especially valuable in scenarios where data is limited or where prior information is crucial for making informed 
                 decisions.
             By allowing for the continuous updating of probabilities, Bayes' Theorem supports a dynamic and flexible approach to statistical analysis,
               which is essential in many real-world applications.'''
```

```python
# Question 19. What is the F-distribution, and when is it used in hypothesis testing ? 
''' Ans=> The F-distribution, also known as the Fisher-Snedecor distribution, is a continuous probability distribution that is often used in hypothesis  
            testing and analysis of variance (ANOVA).
         It is typically used to compare the variability of two population samples or to determine whether two population variances are equal.
         The F-distribution is a right-skewed distribution with a minimum value of 0 and no maximum value.
         It is defined by two parameters, known as the degrees of freedom for the numerator (df1) and the degrees of freedom for the denominator (df2).
         The larger the degrees of freedom, the more the distribution resembles a normal distribution.
        
        The F-distribution is used in hypothesis testing to compare the variances of two samples or to determine whether the means of two or more groups 
          are significantly different from each other.
         In ANOVA, the F-distribution is used to test the null hypothesis that the means of two or more groups are equal. ''' 
```

```python
# Question 20. What is an ANOVA test, and what are its assumptions  ? 
''' Ans=> ANOVA, or Analysis of Variance, is a statistical method used to compare the means of two or more groups by analyzing variance. It helps 
             determine whether the differences observed between group means are statistically significant.
         ANOVA compares the amount of variation between the group means to the amount of variation within each group, using an F-test to assess if the 
             differences are likely due to chance or if they reflect true differences.
        
        The assumptions of ANOVA include:
        
        1.The dependent variable should be measured at the interval or ratio level.
        2.The independent variable should consist of two or more categorical, independent groups.
        3.There should be independence of observations, meaning that there is no relationship between the observations in each group or between the
             groups themselves.
        4.The data should be normally distributed for each group.
        5.There should be no significant outliers.
        6.The variances within each group should be approximately equal (homogeneity of variances).
         This can be tested using Levene’s test for homogeneity of variances
```

```python
# Question 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution ? 

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x: 0 or 1 (possible outcomes)
    - p: probability of success (0 ≤ p ≤ 1)

    Returns:
    - Probability P(X = x)
    """
    if x not in [0, 1]:
        return 0
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")
    
    return p if x == 1 else (1 - p)

print(bernoulli_pdf(1, 0.7))  # Output: 0.7
print(bernoulli_pdf(0, 0.7))  # Output: 0.3
print(bernoulli_pdf(2, 0.7))  # Output: 0
```

```python
# Question 9. Write a Python function to calculate Z-scores from a dataset and plot them. 

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores for a dataset and plot them.
    
    Parameters:
    - data: List or NumPy array of numerical values
    """
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero; Z-scores cannot be calculated.")

    z_scores = (data - mean) / std

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.title("Z-Scores of Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()

    return z_scores

# Sample dataset
data = np.array([10, 12, 23, 23, 16, 23, 21, 16])

# Calculate and plot Z-scores
z = calculate_and_plot_zscores(data)
print("Z-scores:", z)
```

```python
# Question 10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt

# Set parameters
population_size = 100000
sample_size = 50
num_samples = 1000

# 1. Create a non-normal population (Exponential distribution)
population = np.random.exponential(scale=2.0, size=population_size)

# 2. Take repeated samples and compute their means
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# 3. Plot the original distribution and the distribution of sample means
plt.figure(figsize=(14, 6))

# Plot population distribution
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='orange', edgecolor='black', density=True)
plt.title("Original Non-Normal Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Density")

# Plot sample means distribution
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title(f"Sampling Distribution of Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
# Question 15. Implement hypothesis testing using Z-statistics for a sample dataset. 

import numpy as np
from scipy.stats import norm

# Known population parameters
mu = 100       # Population mean
sigma = 15     # Population standard deviation

# Sample data
sample = np.random.normal(loc=104, scale=15, size=50)  # simulate sample
sample_mean = np.mean(sample)
n = len(sample)

# Z-test statistic
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# One-tailed test (right-tail)
p_value = 1 - norm.cdf(z)

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (sample mean is significantly greater than population mean).")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Question 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    
    # Z-test statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    
    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))
    
    # Print results
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.4f}")
    print(f"P-value = {p_value:.4f}")
    
    if p_value < alpha:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

# 🔍 Example usage:
sample_data = [51, 49, 52, 48, 50, 53, 49, 47, 52, 51]  # Sample from unknown distribution
population_mean = 50  # Known population mean
population_std = 2    # Known population standard deviation

z_test(sample_data, population_mean, population_std)
```

```python
# Question 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing. 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, alpha=0.05, n=30):
    """
    Visualizes Type I and Type II errors using two normal distributions.
    
    mu0: mean under null hypothesis (H0)
    mu1: mean under alternative hypothesis (H1)
    sigma: population standard deviation
    alpha: significance level (Type I error rate)
    n: sample size
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for right-tailed test
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se

    # X range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions
    h0_pdf = norm.pdf(x, mu0, se)
    h1_pdf = norm.pdf(x, mu1, se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, h0_pdf, label='H₀ Distribution (μ=%.1f)' % mu0, color='blue')
    plt.plot(x, h1_pdf, label='H₁ Distribution (μ=%.1f)' % mu1, color='green')

    # Shade Type I error region (alpha)
    x_alpha = x[x >= x_critical]
    plt.fill_between(x_alpha, norm.pdf(x_alpha, mu0, se), color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II error region (beta)
    x_beta = x[x < x_critical]
    plt.fill_between(x_beta, norm.pdf(x_beta, mu1, se), color='orange', alpha=0.4, label='Type II Error (β)')

    plt.axvline(x_critical, color='black', linestyle='--', label='Critical Value')

    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_type1_type2(mu0=100, mu1=105, sigma=15, alpha=0.05, n=36)
```

```python
# Question 8 .Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy.stats import ttest_1samp, norm

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=100, scale=15, size=30)  # population mean ~100

# Known population parameters
population_mean = 100
population_std = 15  # Known std for Z-test

# Step 2: Perform Z-test
sample_mean = np.mean(sample)
n = len(sample)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 3: Perform T-test
t_stat, t_p_value = ttest_1samp(sample, population_mean)

# Step 4: Print and compare results
print("=== Z-Test ===")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {z_p_value:.4f}\n")

print("=== T-Test ===")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {t_p_value:.4f}\n")

# Step 5: Interpretation
alpha = 0.05
if z_p_value < alpha:
    print("Z-Test Result: Reject null hypothesis (significant difference).")
else:
    print("Z-Test Result: Fail to reject null hypothesis.")

if t_p_value < alpha:
    print("T-Test Result: Reject null hypothesis (significant difference).")
else:
    print("T-Test Result: Fail to reject null hypothesis.")
```

```python
# Question 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance. 

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default: 0.95)

    Returns:
        (lower_bound, upper_bound)
    """
    sample = np.array(data)
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return round(lower_bound, 4), round(upper_bound, 4)
    
sample_data = [48, 51, 50, 49, 52, 53, 50, 48, 49, 51]
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")


''' 
 Significance of Confidence Interval:
Gives a range estimate instead of a single number.

Indicates reliability: a wider interval = more uncertainty.

A 95% CI means that if you repeated the sampling process 100 times, the true population mean would fall within the interval 95 times.''' 
```

```python
# Question 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the given sample data and confidence level.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default: 0.95)

    Returns:
        margin_of_error (float)
    """
    sample = np.array(data)
    n = len(sample)
    sample_std = np.std(sample, ddof=1)  # sample standard deviation
    std_err = sample_std / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    moe = t_critical * std_err
    return round(moe, 4)

sample_data = [48, 51, 50, 49, 52, 53, 50, 48, 49, 51]
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): {moe}")
```

```python
# Question 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process. 

''' Let’s take a medical test example:

Disease prevalence (prior): 1% → P(Disease) = 0.01

Test sensitivity: 99% → P(Positive | Disease) = 0.99

Test false positive rate: 5% → P(Positive | No Disease) = 0.05 '''

def bayes_theorem(p_h, p_e_given_h, p_e_given_not_h):
    """
    Apply Bayes' Theorem:
    P(H|E) = [P(E|H) * P(H)] / P(E)

    Parameters:
    p_h: Prior probability of H (P(H))
    p_e_given_h: Likelihood (P(E|H))
    p_e_given_not_h: P(E|~H)

    Returns:
    Posterior probability P(H|E)
    """
    p_not_h = 1 - p_h
    p_e = p_e_given_h * p_h + p_e_given_not_h * p_not_h  # Total probability of evidence
    p_h_given_e = (p_e_given_h * p_h) / p_e
    return round(p_h_given_e, 4)

# Given values
prior = 0.01               # P(Disease)
likelihood = 0.99          # P(Positive | Disease)
false_positive = 0.05      # P(Positive | No Disease)

posterior = bayes_theorem(prior, likelihood, false_positive)
print(f"Probability of having disease given a positive test: {posterior}")
```

```python
# Question 12. Perform a Chi-square test for independence between two categorical variables in Python. 

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Example data: a contingency table
# Rows: Gender (Male, Female)
# Columns: Preference (Product A, Product B)
data = [[30, 10],  # Males
        [20, 40]]  # Females

# Create DataFrame for clarity
df = pd.DataFrame(data, columns=["Product A", "Product B"], index=["Male", "Female"])
print("Contingency Table:")
print(df)

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(df)

# Output results
print("\nChi-square Statistic:", round(chi2, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))
print("Expected Frequencies:\n", pd.DataFrame(expected, columns=df.columns, index=df.index))

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nResult: Reject null hypothesis (variables are dependent).")
else:
    print("\nResult: Fail to reject null hypothesis (variables are independent).")
```

```python
# Question 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution. 

from scipy.stats import chisquare
import numpy as np

# Observed frequencies (e.g., outcomes from a die rolled 60 times)
observed = np.array([8, 9, 10, 11, 12, 10])

# Expected frequencies (if die is fair, each face should occur 10 times)
expected = np.array([10, 10, 10, 10, 10, 10])

# Perform chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Chi-square Statistic:", round(chi2_stat, 4))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (data does NOT fit the expected distribution).")
else:
    print("Result: Fail to reject the null hypothesis (data fits the expected distribution).")
```
